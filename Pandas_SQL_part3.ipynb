## Pandas SQL - part 3

```python
import numpy as np 
import pandas as pd 
import json
import sqlite3 as sq3
```

```python
con = sq3.connect("new_movies.db")
```

```python
pd.read_sql("SELECT * \
            FROM Prod", con)
```

```python
df = pd.read_sql("SELECT Prod.id, Prod.prod_name, Movies.revenue, Movies.title \
            FROM Prod \
            LEFT JOIN Movies \
            ON Prod.id=Movies.id \
            WHERE Movies.title LIKE'Titan%'",con)
df
```

```python
df.groupby('prod_name').aggregate({'revenue':'sum'})
```

```python
df.groupby(['prod_name', 'title']).mean()
```

```python
df2 = pd.read_sql("SELECT Genres.id, Genres.genre_name, Movies.revenue, Movies.title \
                    FROM Genres \
                    LEFT JOIN Movies \
                    ON Genres.id=Movies.id", con)
```

```python
df2
```

```python
df2.groupby('genre_name').revenue.sum()
```

```python
df2.groupby('genre_name').aggregate({'revenue':'sum'}).sort_values(by='revenue', ascending=False)
```

```python
#find genre_name for Frozen
pd.read_sql("SELECT Genre_name, Movies.title \
            FROM Movies \
            LEFT JOIN Genres \
            ON Movies.id = Genres.id \
            WHERE Movies.title LIKE 'Frozen%'", con)
```
