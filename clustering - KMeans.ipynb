## Clustering - K Means

- iris dataset: sepal vs petal length determine the species 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set()
```

```python
pd.read_csv('iris-dataset.csv')
```

```python
data = pd.read_csv('iris-dataset.csv')
data
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'])
plt.show()
```

```python
#use cluster

from sklearn.cluster import KMeans
```

```python
knn = KMeans(2)
x = data.copy()
knn.fit(x)
cluster = data.copy()
cluster['cluster'] = knn.fit_predict(x)
plt.scatter(cluster['sepal_length'], cluster['sepal_width'], c = cluster['cluster'], cmap='rainbow')
plt.show()
```

```python
wcss = []
for i in range(1,10):
    knn = KMeans(i)
    x = data.copy()
    knn.fit(x)
    wcss.append(knn.inertia_)
```

```python
wcss
```

```python
plt.plot(wcss)
```

```python
#scaled

from sklearn import preprocessing
```

```python
x = data.copy()
x_scaled = preprocessing.scale(x)
```

```python
wcss = []
for i in range(1,10):
    knn = KMeans(i)
    x = data.copy()
    x_scaled = preprocessing.scale(x)
    knn.fit(x_scaled)
    wcss.append(knn.inertia_)
```

```python
plt.plot(wcss)
```

```python
def plot_cluster(df, n):
    x = df.copy()
    x_scaled = preprocessing.scale(x)
    knn = KMeans(n)
    knn.fit(x_scaled)
    cluster = df.copy()
    cluster['cluster'] = knn.fit_predict(x_scaled)
    plt.scatter(cluster['sepal_length'], cluster['sepal_width'], c=cluster['cluster'], cmap='rainbow')
    plt.show()
    
```

```python
plot_cluster(data, 2)
```

```python
plot_cluster(data, 3)
```
