## import and merge multiple text files 

```python
import numpy as np
import pandas as pd
```

```python
#import files by year names
pd.read_csv("names\yob1990.txt")
```

```python
# of files 
# create a list - yob from 1895 to 2019

years = list(range(1895,2020))
years
```

```python
#header = 0
#names = ['Name','Gender','Count']

yob_years=[]
for i in years:
    data = pd.read_csv("names\yob{}.txt".format(i), header=None, names=['Name','Gender','Count'])
    yob_years.append(data)
```

```python
yob_years
```

```python
#combine all of the data from years 1895 to 2019

pd.concat(yob_years, keys=years, names=['Year'])
```

```python
pd.concat(yob_years, keys=years, names=['Year']).droplevel(-1)
```

```python
df = pd.concat(yob_years, keys=years, names=['Year']).droplevel(-1).reset_index()
df.info()
```

```python
#save combined data 
df.to_csv("us_baby_names.csv", index=False)
```

## utilize glob module 

```python
from glob import glob #all file /path names by specificying pattern
```

```python
pd.read_csv("namestates/AK.TXT", header=None, names=['State','Gender','Year','Name','Count'])
```

```python
glob("namestates\\*.TXT")
```

```python
pd.read_csv('namestates\\WY.TXT')
```

```python
filenames = glob("namestates\\*.TXT")
```

```python
len(filenames)
```

```python
file_list=[]
for i in filenames:
    df =pd.read_csv(i, header=None, names=['State','Gender','Year','Name','Count'])
    file_list.append(df)
```

```python
file_list
```

```python
df1 = pd.concat(file_list, axis=0, ignore_index=True)
df1.head()
```

```python
df1.to_csv("baby_names_states.csv", index=False)
```
