## Pandas SQL

```python
import pandas as pd 
import sqlite3 as sq3
```

```python
con = sq3.connect("new_movies.db")
```

```python
pd.read_sql("SELECT * \
            FROM Movies", con)
```

```python
pd.read_sql("SELECT * \
            FROM Movies \
            WHERE id ='597'", con)
```

```python
pd.read_sql("SELECT DISTINCT title \
            FROM Movies", con)
```

```python
#DISTINCT - unique 
pd.read_sql("SELECT DISTINCT belongs_to_collection_name \
            FROM Movies", con)
```

```python
pd.read_sql("SELECT id, belongs_to_collection_name \
            FROM Movies \
            WHERE belongs_to_collection_name IS NULL", con)
```

```python
pd.read_sql("SELECT * \
            FROM Movies \
            WHERE revenue > 1500 AND budget < 200 \
            ORDER BY revenue ASC", con)
```

```python
pd.read_sql("SELECT * \
            FROM Movies \
            WHERE title LIKE 'Ave%'", con)
```

```python
##join - movies and votes table

pd.read_sql("SELECT * \
            FROM Movies \
            JOIN Votes \
            ON Movies.id = Votes.id", con)
```

```python
pd.read_sql("SELECT Movies.id, Movies.title, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id = Votes.id \
            WHERE Votes.vote_average > 7 \
            ORDER BY Votes.vote_average ASC", con, index_col='id')
```

```python
con.close()
```
