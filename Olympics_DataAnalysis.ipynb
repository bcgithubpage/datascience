```python
dictionary.info()
```

```python
dictionary.rename(columns={'GDP per Capita':'GDP'}, inplace=True)
```

```python
dictionary.loc[dictionary.isnull().any(axis=1)] #any row with a missing value
```

```python
#combine winter and summer olympics
#map country based on the country 'code'

#use keys - winter and summer to differentiate 

pd.concat([winter,summer], axis=0, keys=['winter','summer'], names=['season']).droplevel(-1).reset_index()
```

```python
df_olympics = pd.concat([winter, summer], axis=0, keys=['Winter','Summer'], names=['Season']).droplevel(-1).reset_index()
df_olympics.head()
```

```python
#map the code from dictionary
#first, rename the df_olympics Country to 'Code'

df_olympics.rename(columns={'Country':'Code'}, inplace=True)
```

```python
df_olympics.shape
```

```python
df_olympics = df_olympics.merge(dictionary.iloc[:, :2], how='left', on='Code')
df_olympics.head()
```

```python
df_olympics.isnull().sum()
```

```python
df_olympics.loc[df_olympics.isnull().any(axis=1)][['Country','Code']]
```

```python
#obtain index of the missing Country column
missing_c_index = df_olympics.loc[df_olympics['Country'].isnull()].index
missing_c_index
```

```python
df_olympics.loc[df_olympics.isnull().any(axis=1)]['Code'].value_counts()
```

```python
#old country codes 

old_c = df_olympics.loc[df_olympics.isnull().any(axis=1)]['Code'].value_counts().index
old_c
```

```python
data=['Soviet Union', 'East Germany','Romania', 'West Germany', 'Czechoslovakia', 'Yugoslavia','Unified Team', 'Unified Team of Germany', 'Mixed Teams', 'Serbia', 'Australasia', 'Russian Empire', 'Montenegro','Trinidad and Tobago', 'Bohemia', 'West Indies Federation', 'Singapore', 'Independent Olympic Participants']
```

```python
#map missing country names that have old country code with old country names
#index using the old country code
#map to data listed above

mapper = pd.Series(index=old_c, name='Country', data=data)
mapper
```

```python
#use mapper to map the "Code" to the missing Country by the missing_c_index

df_olympics.loc[missing_c_index, 'Code'].map(mapper)
```

```python
#identify the missing Country 
#fillna of the missing Country using the Code that maps to mapper

df_olympics.Country.fillna(df_olympics.Code.map(mapper), inplace=True)
```

```python
df_olympics.isnull().sum()
```

```python
df_olympics.loc[df_olympics.isnull().any(axis=1)]
```

```python
#athlete name is also pending for 3
#could be related to doping 
#delete these rows - either by 'Code' or 'Country'

df_olympics.dropna(subset=['Country'], inplace=True)
```

```python
df_olympics.isnull().sum()
```

```python
df_olympics.nunique()
```

```python
#if small # of categories, could make sense to change datatype to category, e.g. medal

df_olympics.Medal = df_olympics.Medal.astype('category')
```

```python
df_olympics.info()
```

```python
df_olympics['Medal'].sort_values()
```

```python
#change the order - Bronze, Silver then Gold

df_olympics.Medal.cat.set_categories(['Bronze','Silver','Gold'], ordered=True, inplace=True)
```
