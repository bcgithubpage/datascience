## Import Data via Web Api

- data source: /www.themoviedb.org

refer to data source to obtain correct web path

```python
api_key = 'api_key=2150281ee836fc2bb45a4164cf0de15c'
```

```python
basic_url = "https://api.themoviedb.org/3/movie/{}?{}"
```

```python
#selected 3 random movie ids 
movie_id = [299534, 19995, 140607]
```

```python
json_list = []

for i in movie_id:
    url = basic_url.format(i, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data)
```

```python
#check json_list - normalization 

json_list
```

```python
pd.json_normalize(data=json_list)
```

```python
df= pd.json_normalize(data=json_list)
```

```python
#however, genres couldn't be normalized since its nested 
#need to create another dataframe with genres separated to each movies

df_genres = pd.json_normalize(data=json_list, record_path='genres', record_prefix='genres_', meta=['id','title'])
```
